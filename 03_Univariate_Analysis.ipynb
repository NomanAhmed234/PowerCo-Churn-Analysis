# Notebook 03: Univariate Analysis

## Objective

After confirming the quality and integrity of the PowerCo customer churn dataset in the previous notebook, the next step is to understand each feature individually through **Univariate Analysis**.

Univariate Analysis is the process of analyzing **one variable at a time**. The primary objective is to understand the distribution, characteristics, and behavior of each feature independently, without considering its relationship with other variables.

A thorough understanding of individual features is essential because it helps identify important patterns, unusual observations, potential data quality issues, and preprocessing requirements before exploring relationships between multiple variables or building machine learning models.



## Why is Univariate Analysis Important?

Every feature in a dataset tells a different part of the business story.

For an energy company like PowerCo, individual variables describe different aspects of a customer, such as:

- Customer energy consumption
- Contract information
- Electricity pricing
- Customer profitability
- Product subscriptions
- Forecasted consumption
- Sales channels
- Customer acquisition source

By studying each feature separately, we can answer questions such as:

- How is the data distributed?
- Are there unusual or extreme values?
- Does the feature contain outliers?
- Is the feature highly skewed?
- Which values occur most frequently?
- Does the feature require preprocessing before modeling?

Understanding these characteristics allows us to make informed decisions during feature engineering and model development.



# Types of Features in the Dataset

The PowerCo customer churn dataset contains different types of variables, and each type requires a different analytical approach.

## 1. Numerical Features

Numerical features contain quantitative values that can be measured or counted.

Examples include:

- Annual electricity consumption
- Monthly consumption
- Net margin
- Power subscription
- Forecasted consumption
- Electricity prices

Since numerical data consists of continuous or discrete numbers, we analyze these variables using statistical measures and distribution-based visualizations.

For each numerical feature, we will examine:

- Descriptive statistics
- Distribution
- Histogram
- Density (KDE) Plot
- Box Plot
- Skewness
- Kurtosis
- Outliers
- Business interpretation
- Machine learning interpretation

These analyses help us understand the spread, variability, central tendency, and overall distribution of each numerical variable.



## 2. Categorical Features

Categorical features represent groups or labels rather than numerical quantities.

Examples include:

- Sales Channel
- Customer Origin
- Gas Customer Status
- Business Activity

Since categorical variables represent categories instead of numerical measurements, statistical measures such as mean or standard deviation are not meaningful.

Instead, we analyze these features using:

- Frequency tables
- Percentage distribution
- Bar charts
- Pie charts
- Cardinality analysis
- Mode
- Business interpretation
- Machine learning interpretation

These analyses help identify the most common customer groups, dominant business categories, and the diversity of categorical values.



## 3. Date Features

Date variables describe events that occurred over time.

Examples include:

- Contract activation date
- Contract end date
- Product modification date
- Contract renewal date

Unlike numerical or categorical variables, date features provide temporal information that helps us understand customer lifecycles and business operations.

For each date feature, we will examine:

- Earliest and latest dates
- Date ranges
- Timeline distributions
- Customer activity over time
- Business interpretation

These analyses provide insights into customer acquisition patterns, contract durations, and renewal behavior.



# Why Do Different Data Types Require Different Analyses?

Each type of data contains different information and therefore requires specialized analytical techniques.

| Feature Type | Nature of Data | Appropriate Analysis | Purpose |
|--------------|----------------|----------------------|---------|
| Numerical | Quantitative values | Statistics, Histograms, KDE, Boxplots, Skewness, Outliers | Understand distribution, variability, and unusual observations |
| Categorical | Labels or categories | Frequency Tables, Bar Charts, Pie Charts, Cardinality | Understand category distribution and dominant groups |
| Date | Time-based information | Timeline Analysis, Date Ranges, Temporal Trends | Understand customer lifecycle and business events |

Using the same analysis for every type of variable would not provide meaningful insights. For example, calculating the mean of a sales channel or creating a histogram of contract dates would not accurately represent the underlying information.

Selecting the appropriate analytical method for each data type ensures that the results are statistically valid, easy to interpret, and relevant to the business problem.
